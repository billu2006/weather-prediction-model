```python
import pandas as pd

weather = pd.read_csv("weather.csv", index_col = "DATE")#reads in the csv file containing weather data
```

```python
weather
```

```python
null_pct = weather.apply(pd.isnull).sum()/weather.shape[0]   #performs the calculation: number of null values / number of rows = percentage of null values
```

```python
null_pct
```

```python
#missing data could be for a variety of reasons, e.g: someone forgot to write the data that day, a sensor could have malfunctioned or the sensor just hadnt been installed yet.
```

```python
#clean data by removing any columns where null_pct is too low

valid_columns = weather.columns[null_pct < 0.05]
```

```python
valid_columns
```

```python
#change dataset to where it only contains the valid columns above

weather = weather[valid_columns].copy() # .copy() to avoid copy warnings later in the code
```

```python
weather.columns = weather.columns.str.lower() #lowercasing column names for convenience
```

```python
#filling in missing data

weather = weather.ffill() #.ffill() looks for a missing peice of data and fills it in with the last non-missing peice of data

#this is a reasonable assumption especially for feilds such as snow depth because if yesterday the snow depth was 0.0 then today it is also most likely 0.0
```

```python
#create a target column

weather["target"] = weather.shift(-1)["tmax"] #takes the tmax for the day after and puts it into the target
```

```python
#most recent date doesnt have a target so use ffill to fill in the gap

weather = weather.ffill()
```

```python
#apply a ridge regression model

from sklearn.linear_model import Ridge

rr = Ridge(alpha=0.1)
```

```python
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])]
```

```python
predictors
```

```python
#use time series cross validation so we dont use future data to predict past date

def backtest(weather, model, predictors, start = 3650, step = 90): # start makes sure we take 10 years of data before starting to make predictions and step ensures that we make a set of predictions every 90 days
    all_predictions = []

    for i in range(start, weather.shape[0], step): #each loop generates predictions for 90 days
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)

        combined.columns = ["actual", "prediction"]

        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions, axis=0)
```

```python
predictions = backtest(weather, rr, predictors)
```

```python
predictions
```

```python
#to see how accurate the predictions are we need to generate an accuraccy metric (in this case we will use mean absolute error)

from sklearn.metrics import mean_absolute_error

mean_absolute_error(predictions["actual"], predictions["prediction"]) # takes all of the differences and finds the average
```

```python
# on average we were 5.13932667966084 F off of the correct temp
```

```python
#try to improve accuracy

def pct_diff(old, new):
    return (new - old) / old

def compute_rolling(weather, horizon, col):
    label = f"rolling_{horizon}_{col}" #upon reasearch this is the best way to make a dynamic string

    weather[label] = weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather

rolling_horizons = [3, 14]

for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, horizon, col)
```

```python
#missing values as we dont have a 14 day rolling average for these sets of data
```

```python
weather = weather.iloc[14:, :] #removes the first 14 rows because they are incomplete
```

```python
#some pct collumns are still null because we are dividing by 0 so we can just fill in those values to be 0

weather = weather.fillna(0)
```

```python
weather
```

```python
def expand_mean(df):#returns the row in the dataframe and all of the previous rows and returns the mean
    return df.expanding(1).mean()

for col in ["tmax", "tmin", "prcp"]:
    weather[f"month_avg_{col}"] = weather[col].groupby(weather.index.month, group_keys = False).apply(expand_mean)
    weather[f"day_avg_{col}"] = weather[col].groupby(weather.index.day_of_year, group_keys = False).apply(expand_mean)
```
